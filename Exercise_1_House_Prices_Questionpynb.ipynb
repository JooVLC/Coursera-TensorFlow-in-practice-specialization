```python
import tensorflow as tf
import numpy as np
from tensorflow import keras

print(tf.__version__)
```

```python
def housing_price_in_hundreds_of_thousands(qtd_of_bedrooms):
  base_house_price_without_bedroom = 5
  price_per_bedroom = 5
  return base_house_price_without_bedroom + (price_per_bedroom * qtd_of_bedrooms)
```

```python
simple_model = keras.Sequential(layers=[keras.layers.Dense(units=1, input_shape=[1])])
simple_model.compile(optimizer="sgd", loss="mean_squared_error")
```

```python
xs_qtd_of_bedrooms = np.array([1, 2, 5, 10, 0, 4, 9, 6], dtype=np.uint)
ys_house_price_in_hundreds_of_thousands = housing_price_in_hundreds_of_thousands(xs_qtd_of_bedrooms)

print(xs_qtd_of_bedrooms)
print(ys_house_price_in_hundreds_of_thousands)
```

```python
simple_model.fit(xs_qtd_of_bedrooms, ys_house_price_in_hundreds_of_thousands, epochs=1000)
```

```python
print(simple_model.predict([7]))
```
